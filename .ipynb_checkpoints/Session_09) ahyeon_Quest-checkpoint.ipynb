## 몇개의 주성분이 적당할까요?

```python
from sklearn.decomposition import PCA

# PCA
pca = PCA().fit(X_scaled)

# 몇개의 주성분이 적당할까? 
print('explained variance ratio : ', pca.explained_variance_ratio_)
# PCA().fit(X_scaled).explained_variance_ratio_ 와 같음
print('cumulative variance : ', np.cumsum(pca.explained_variance_ratio_)) 
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
#14개 중 6개의 누적설명분산량이 85%이므로 주성분을 6개로 결정
```

## PCA 를 진행해봅시다. 

```python
pca = PCA(n_components=6)

X_pca=pca.fit_transform(X_scaled)
# PCA(n_components=2).fit_transform(x_scaled) 와 같음
print(X_pca.shape)
```

```python
# PCA 변환된 데이터의 컬럼명을 각각 pca_1, pca_2, ... pca_6로 명명
pca_columns=['pca_1','pca_2', 'pca_3', 'pca_4', 'pca_5', 'pca_6']
wine_pca = pd.DataFrame(X_pca,columns=pca_columns)
wine_pca['Customer_Segment']=df.Customer_Segment
wine_pca.head(3)
```

## 주성분의 loadings를 찾아봅시다.
세션에서는 다루지 않았던 내용입니다. 구글링해서 도전해보세요!

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6'], index=df.columns)
loadings
# The loadings can be understood as the weights for each original variable when calculating the principal component.
```

# *Hierarchical clustering
실습에서 사용한 iris 데이터에서 linkage 를 달리하여 dendrogram 을 그려보세요

```python
from sklearn import datasets
iris = datasets.load_iris()

labels = pd.DataFrame(iris.target)
labels.columns=['labels']
data = pd.DataFrame(iris.data)
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
data = pd.concat([data,labels],axis=1)

data.head()
```

```python
#교육세션에서는 average linkage 사용해서 complete과 single을 사용
from scipy.cluster.hierarchy import linkage, dendrogram

mergings = linkage(data,method='complete') #method =linkage설정하는것
#method에는 complete(최장연결), single(최단연결)도 있음

# Plot the dendrogram
plt.figure(figsize=(40,20))
dendrogram(mergings,
           labels = labels.values,
           leaf_rotation=90,
           leaf_font_size=20,
)
plt.show()
```

```python
mergings = linkage(data,method='single') #method =linkage설정하는것
#method에는 complete(최장연결), single(최단연결)도 있음

# Plot the dendrogram
plt.figure(figsize=(40,20))
dendrogram(mergings,
           labels = labels.values,
           leaf_rotation=90,
           leaf_font_size=20,
)
plt.show()
```
