```python
# 전처리 전체 작업
```

```python
def text_cleaning(review):
    review = BeautifulSoup(review, "html.parser").get_text()
    only_eng = re.sub("[^a-zA-Z]", " ", review)
    no_cap = only_eng.lower().split()
    stops = set(stopwords.words('english'))
    no_stops = [word for word in no_cap if not word in stops]
    stemmer = nltk.stem.SnowballStemmer('english')
    stem_word = [stemmer.stem(word) for word in no_stops]
    return ' '.join(stem_word)
```

```python
clean_review = []
for rv in hotel_df['review_body']:
    clean_review.append(text_cleaning(rv))
```

```python
review_df = pd.DataFrame({'review': clean_review})
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_review)
text_sequences = tokenizer.texts_to_sequences(clean_review)
```

```python
MAX_SEQUENCE_LENGTH = 174 

hotel_input = pad_sequences(text_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post')

print('shape of data: ', hotel_input.shape)
```

```python
INPUT_DATA = 'hotel_input.npy'
HOTEL_DATA = 'hotel_clean.csv'
np.save(open( INPUT_DATA , 'wb'), hotel_input)
review_df.to_csv(HOTEL_DATA, index = False)
```
