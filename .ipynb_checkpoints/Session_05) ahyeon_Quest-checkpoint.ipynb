```python
daily_boxoffice_1 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_0"})
daily_boxoffice_2 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_130"})
daily_boxoffice_3 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_260"})
daily_boxoffice_4 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_377"})
daily_boxoffice_5 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_504"})
daily_boxoffice_6 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_617"})
daily_boxoffice_7 = parsed_daily_boxoffice_html.find("tbody", {"id":"tbody_722"})
```

```python
movie_list_1 = daily_boxoffice_1.find_all("tr")
movie_list_2 = daily_boxoffice_2.find_all("tr")
movie_list_3 = daily_boxoffice_3.find_all("tr")
movie_list_4 = daily_boxoffice_4.find_all("tr")
movie_list_5 = daily_boxoffice_5.find_all("tr")
movie_list_6 = daily_boxoffice_6.find_all("tr")
movie_list_7 = daily_boxoffice_7.find_all("tr")
movie_list=[]
movie_list = [movie_list_1, movie_list_2, movie_list_3, movie_list_4, movie_list_5, movie_list_6, movie_list_7]
```

```python
import pandas as pd

daily_movies = []

for i in range(7):
    daily_movies.append({'영화명' : [],
'순위 변동' : [],
'개봉일' : [],
'매출액' : [],
'매출액점유율' : [],
'관객수' : [],
'누적관객수' : [],
'스크린수' : [],
'상영횟수' : []})

i = 0
for lists in movie_list:
    for movie in lists:
        title_rank = movie.find("td").find_next_sibling("td").text
        daily_movies[i]['영화명'].append(title_rank[:-5].strip())
        daily_movies[i]['순위 변동'].append(title_rank[-4:].strip())
        daily_movies[i]['개봉일'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['매출액'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['매출액점유율'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['관객수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['누적관객수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['스크린수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
        daily_movies[i]['상영횟수'].append(movie.find("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").find_next_sibling("td").text.strip())
    i += 1
```

```python
daily_boxoffice_0 = pd.DataFrame(daily_movies[0])
daily_boxoffice_1 = pd.DataFrame(daily_movies[1])
daily_boxoffice_2 = pd.DataFrame(daily_movies[2])
daily_boxoffice_3 = pd.DataFrame(daily_movies[3])
daily_boxoffice_4 = pd.DataFrame(daily_movies[4])
daily_boxoffice_5 = pd.DataFrame(daily_movies[5])
daily_boxoffice_6 = pd.DataFrame(daily_movies[6])

def change_rank(rank):
    if rank == "동일":
        rank = 0
    elif rank[-2:] == "하락":
        rank = int(rank[:-2]) * -1
    elif rank[-2:] == "상승":
        rank = int(rank[:-2]) * 1
    return rank

daily_boxoffice_0['순위 변동'] = daily_boxoffice_0['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_1['순위 변동'] = daily_boxoffice_1['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_2['순위 변동'] = daily_boxoffice_2['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_3['순위 변동'] = daily_boxoffice_3['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_4['순위 변동'] = daily_boxoffice_4['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_5['순위 변동'] = daily_boxoffice_5['순위 변동'].map(lambda x: change_rank(x))
daily_boxoffice_6['순위 변동'] = daily_boxoffice_6['순위 변동'].map(lambda x: change_rank(x))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import sys


import warnings
warnings.filterwarnings('ignore')

# notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt 
import platform                

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

if platform.system() == 'Darwin': #맥os 사용자의 경우에
    plt.style.use('seaborn-darkgrid') 
    rc('font', family = 'AppleGothic')
    
```

```python
df=pd.read_csv("nyc_citibike.csv",encoding='euc-kr')
df.head()
```

```python
df['start_station_id']=df['start_station_id'].astype(str)
df['end_station_id']=df['end_station_id'].astype(str)
df['bike_id']=df['bike_id'].astype(str)
```

```python
df['trip_duration_min']=df['trip_duration'] /60
```

```python
cut_point = df["trip_duration_min"].quantile(0.99) # 상위 99% 값을 cut_point로 지정
df_cut=df[df['trip_duration_min'] < cut_point]
```

```python
df_cut['trip_duration_min'].mean()
```

```python
df_cut.groupby('bike_id').mean()
```

```python
#주행시간의 평균이 상위 20위인 bike_id만
df_cut['trip_duration_min']
id_df=df_cut.groupby('bike_id').mean().sort_values(by='trip_duration_min', ascending=False)
top_list=id_df.iloc[:20].index
df_top=df_cut.loc[df_cut['bike_id'].isin(top_list), :]
```

```python
df_top=df_cut.loc[df_cut['bike_id'].isin(top_list), :]
df_top.reset_index()
```

```python
plt.figure(figsize=(10,6))
box=sns.boxplot(x='bike_id',y='trip_duration_min',data=df_top)
box.set_xlabel("bike id")
box.set_ylabel("주행시간")
plt.show()
# 생각보다 상위 20위의 평균주행시간의 bike_id들 중 고유한 값? 이 많아서 Boxplot이 선으로 나타난 점이 많았다.
# 누적으로 했으면 나았을까..?
```
